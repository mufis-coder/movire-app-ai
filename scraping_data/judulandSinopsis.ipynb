```python
import scrapy
from scrapy.crawler import CrawlerProcess

class DC_Description_Spider(scrapy.Spider):
    name = "dc_chapter_spider"
    
    def start_requests(self):
        url_page = url_film
        for url_l in url_page:    
            yield scrapy.Request(url = url_l, callback = self.parse_awalpage)
            
    def parse_awalpage(self, response):
#         print(response.text)
        film_title = response.xpath('//h1[contains(@class,"section__film-info__entitle section__title")]/text()')
        film_title_ext = film_title.extract_first().strip()
        
        film_sinopsis = response.xpath('//div[contains(@class,"section__film-info__description-body after-none")]//p/text()')
        film_sinopsis = film_sinopsis.extract()
        
        dawa = len(film_sinopsis)
        if dawa > 1:
#             print('FILM LEBIH DARI 1 SINOP: ', film_title_ext)
            words = ""
            for sin in film_sinopsis:
                words = words + " " + sin
            words = words[1:]
            film_sinopsis = words
#             print(film_sinopsis)
        else:
            film_sinopsis = film_sinopsis[0]
        
        list_judul.append(film_title_ext)
        list_sinopsis.append(film_sinopsis)
        print('Jmlah judul', len(list_judul))
        print('Jmlah sinopsis', len(list_sinopsis))
```

```python
list_judul = list()
list_sinopsis = list()
process = CrawlerProcess()
process.crawl(DC_Description_Spider)
process.start()
```

```python
dict = {'Judul': list_judul, 'Sinopsis':list_sinopsis}   
       
df = pd.DataFrame(dict)  
    
# saving the dataframe  
df.to_csv('Data_film_bindo.csv', index=False) 
```
